```python
#DATA
def extract_ingredients(row):
    ingredients = row['extendedIngredients']
    return [{'id': ingredient['id'], 'nameClean': ingredient['nameClean']} for ingredient in ingredients]

# Apply the function to each row and create a new column with the result
data['listIngredients'] = data.apply(extract_ingredients, axis=1)

# Display the DataFrame with the updated colum
data['listIngredients'].tolist()
```

```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
# Function to group similar ingredients
def group_similar_ingredients(ingredients_list):
    grouped_ingredients = {}

    for ingredient in ingredients_list:
        found_group = False
        for group in grouped_ingredients.keys():
            for group_member in grouped_ingredients[group]:
                if fuzz.partial_ratio(ingredient['nameClean'], group_member['nameClean']) > 80:  # Adjust threshold as needed
                    grouped_ingredients[group].append(ingredient)
                    found_group = True
                    break
            if found_group:
                break
        if not found_group:
            grouped_ingredients[ingredient['nameClean']] = [ingredient]

    return list(grouped_ingredients.values())

# Apply the function to each row and create a new column with the result
data['groupedIngredients'] = data['listIngredients'].apply(group_similar_ingredients)

# Display the DataFrame with the updated column
data['groupedIngredients'].tolist()
```

```python
from fuzzywuzzy import fuzz

# Function to group similar ingredients
def group_similar_ingredients(ingredients_list):
    grouped_ingredients = {}

    for ingredient in ingredients_list:
        found_group = False
        for group in grouped_ingredients.keys():
            for group_member in grouped_ingredients[group]:
                if fuzz.partial_ratio(ingredient['nameClean'], group_member['nameClean']) > 80:  # Adjust threshold as needed
                    grouped_ingredients[group].append(ingredient)
                    found_group = True
                    break
            if found_group:
                break
        if not found_group:
            grouped_ingredients[ingredient['nameClean']] = [ingredient]

    # Flatten the dictionary values to get a list of grouped ingredients
    return list(grouped_ingredients.values())

# Apply the function to each row and create a new column with the result
data['groupedIngredients'] = data['listIngredients'].apply(group_similar_ingredients)
```

## group similar ingredients

```python
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


# Extract unique ingredients from the DataFrame
ingredients = set()

for index, row in data.iterrows():
    ingredients_list = row['extendedIngredients']
    for ingredient in ingredients_list:
        ingredients.add(ingredient['nameClean'])

# Tokenization and lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_ingredients = [' '.join([lemmatizer.lemmatize(word) for word in word_tokenize(ingredient)]) for ingredient in ingredients]

# Vectorize ingredients using TF-IDF
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(lemmatized_ingredients)

# Calculate cosine similarity between ingredients
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Threshold for considering ingredients as similar
threshold = 0.8

# Group similar ingredients
similar_groups = {}

for i in range(len(lemmatized_ingredients)):
    ingredient = lemmatized_ingredients[i]
    similar_indices = [idx for idx, sim in enumerate(cosine_sim[i]) if sim > threshold and idx != i]
    if similar_indices:
        similar_ingredients = [ingredients[idx] for idx in similar_indices]
        similar_ingredients.append(ingredients[i])
        similar_groups[ingredient] = similar_ingredients

# Display similar ingredient groups
for group in similar_groups.values():
    print(group)
```
